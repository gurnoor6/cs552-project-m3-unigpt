```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import numpy as np
```

# 1 electra normal

```python
with open('./results/result_electra.json') as f:
    eval_results = json.load(f)
electra_df = pd.DataFrame(eval_results)
```

```python
rewards_1_electra = electra_df[electra_df['confidence'] == 1]['predictions'].to_list()
rewards_2_electra = electra_df[electra_df['confidence'] == 2]['predictions'].to_list()
rewards_3_electra = electra_df[electra_df['confidence'] == 3]['predictions'].to_list()
rewards_4_electra = electra_df[electra_df['confidence'] == 4]['predictions'].to_list()
rewards_5_electra = electra_df[electra_df['confidence'] == 5]['predictions'].to_list()
```

```python
# Create the box plots for each confidence level
box_1 = go.Box(y=rewards_1_electra, name="1")
box_2 = go.Box(y=rewards_2_electra, name="2")
box_3 = go.Box(y=rewards_3_electra, name="3")
box_4 = go.Box(y=rewards_4_electra, name="4")
box_5 = go.Box(y=rewards_5_electra, name="5")


# Create the layout for the plot
layout = go.Layout(
    title={
        'text': "Electra Reward",
        'y':0.95,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'},
    title_font=dict(size=30, color='rgb(0, 0, 0)', family="Courier New, monospace"),  # black color
    xaxis=dict(
        title='Confidence', 
        titlefont=dict(size=18, color='rgb(0, 0, 0)', family="Courier New, monospace"),  # black color
        tickfont=dict(size=16, color='rgb(0, 0, 0)', family="Courier New, monospace")),  # black color
    yaxis=dict(
        title='Reward',
        titlefont=dict(size=18, color='rgb(0, 0, 0)', family="Courier New, monospace"),  # black color
        tickfont=dict(size=16, color='rgb(0, 0, 0)', family="Courier New, monospace")),  # black color  
    paper_bgcolor='rgb(255,255,255)',
    plot_bgcolor='rgb(229,229,229)',
    width=12*96,  # Width of the figure in pixels (12 inches * 96 ppi)
    height=8*96,  # Height of the figure in pixels (12 inches * 96 ppi)

)

#refize the plot to 12,12

# Create the figure and add the box plots to it
fig = go.Figure(data=[box_1, box_2, box_3, box_4, box_5], layout=layout)

# Show the figure
fig.show()

#save the figure
fig.write_image("./plots/electra_rewards_by_confidence.png")



```
